# Bar plot in Matplot library

```python
import matplotlib.pyplot as plt 
x=['python','c++','c','java']
y=[20,30,40,50]
c=['r','g','y','b']
plt.bar(x,y,width=0.3,color=c,align='edge',edgecolor='m',linewidth=5,linestyle=':',alpha=0.5,label='wscube',)
plt.legend()
plt.xlabel('programming languages',loc='right')
plt.ylabel('popularity of languages',fontsize=20)
plt.title('languages popularity')
plt.grid(True)
plt.show()
```

# Scatter plot in Matplot library

```python
import matplotlib.pyplot as plt 
x=[1,2,3,5,6]
y=[2,3,4,6,7]
color=['y','g','b','r','m']
plt.scatter(x,y,marker="*",s=800,c=color,edgecolors="m",linewidths=1,linestyle=":",alpha=0.3)
plt.title("scatter")
plt.xlabel("numbers")
plt.show()
```

```python
# Implementation of matplotlib.pyplot.xlabels() 
# function 
   
import numpy as np 
import matplotlib.pyplot as plt 
  
t = np.arange(-180.0, 180.0, 0.1) 
s = np.radians(t)/2.
  
plt.plot(t, s, '-', lw = 2) 
  
plt.xlabel('Longitude') 
plt.ylabel('Latitude') 
plt.title('xlabels() function') 
plt.grid(True) 
  
plt.show() 
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5,6]
y=[40,60,40,60,30,50]
plt.plot(x,y,color="y",linewidth=7)
plt.title("colours scheme")
plt.xlabel("numbers",fontsize=50,color="y")
plt.ylabel("colours")
plt.grid(True)
plt.show()
```
